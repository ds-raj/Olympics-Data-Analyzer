```python
df = df[df['Season'] == 'Summer']
df = df.merge(regions,on = 'NOC',how = 'left')
df.drop_duplicates(inplace = True)
df = pd.concat([df,pd.get_dummies(df['Medal'])],axis = 1)
```

```python
def most_successfull(df,selected_sport):
    temp = df.dropna(subset = ['Medal'])
    selected_sport = 'Overall'
    if selected_sport != 'Overall':
        temp = temp[temp['Sport'] == selected_sport]
    x = temp['Name'].value_counts().reset_index().merge(df,left_on = 'index',right_on = 'Name',how= 'left')[['index','Name_x','region','Sport']].drop_duplicates('index')
    x.rename(columns={'index':'Name','Name_x':'Medals'})
    x.drop('level_0',axis=1)
    return x
```
